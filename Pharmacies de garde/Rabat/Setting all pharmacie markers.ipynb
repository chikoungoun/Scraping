# Set all the Rabat pharmacies as markers on the map

```python
with open('../api_key.txt') as f:
    api_key = f.readline()
    f.close
```

```python
import gmaps
```

```python
gmaps.configure(api_key=api_key)
```

### Opening the csv file of the pharmacie Database

```python
import pandas as pd
```

```python
df = pd.read_csv('pdg_rabat.csv',delimiter=';',index_col=0)
```

```python
df
```

### Set the markers locations

```python
marker_locations = []
for j in df['coordonnee']:
    if j != "NO COORDINATES":
        x,y = j.split(',')
        print(x)
        print(y)
        
        t = (float(x),float(y))
        marker_locations.append(t)
```

```python
marker_locations
```

### Display all the markers on the map

```python
fig = gmaps.figure()
markers = gmaps.marker_layer(marker_locations)
fig.add_layer(markers)
fig
```
