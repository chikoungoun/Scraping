# Combining the data of 2 datasets

```python
import pandas as pd
```

```python
df = pd.read_csv('pdg_rabat.csv',delimiter=';',index_col=0)
```

```python
df
```

## Dataset containing the neighborhoods

```python
dq = pd.read_excel('q.xlsx')
```

```python
dq
```

**Cleaning dataset**

```python
dq.drop(['Unnamed: 0', 'Unnamed: 1'],axis=1,inplace=True)
```

## Merging the dataframes

```python
rs = pd.merge(df,dq,on=['pharmacie','lien','coordonnee'])
```

```python
rs
```

```python
rs.drop(['access_x'],axis=1,inplace=True)
rs.drop(['access_y'],axis=1,inplace=True)
```

```python
rs
```

## Replace the phone nuymber so it contains the first "0"

```python
rs['phones']= rs['phones'].astype(str)
```

```python
for i, row in rs.iterrows():
    row['phones'] = "0"+str(row['phones'])
```
