# Scraping all the Rabat pharmacies from Telecontact

```python
url ="https://www.telecontact.ma/liens/pharmacies/rabat.php&page="
```

```python
from bs4 import BeautifulSoup
```

```python
from time import time
from time import sleep
from random import randint
```

```python
start_time = time()
requests = 0
```

```python
from requests import get
```

```python
liens = []
pharmacies = []
maps = []
coor = []

z = 1
while (z < 17):
    print(" --- ---  PAGE "+str(z)+" --- ---")
    print(url+str(z))
    lien = url+str(z)
    
    response = get(lien)
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    soup = BeautifulSoup(response.text,'html.parser')
    
    results = soup.find_all('article',class_='drs col-xs-12')
    
    for i in results:
        liens.append("https://www.telecontact.ma"+str(i.find('span',id='resultats_h3_span').findAll('a')[1]['href']))
        
        
    for i in results:
        pharmacies.append("Pharmacie "+str(i.find('span',id='resultats_h3_span').findAll('a')[2].text))
        
    maps_t = []    
    for i in results:
        m = i.find('div',class_="col-xs-12 no-margin").find('div',class_="col-xs-12 no-margin").find('a')
    
        if m['title'] == "Plan d'acc�s":
            maps_t.append("https://www.telecontact.ma"+m['href'])
        else:
            maps_t.append("NO ACCESS")
    
    # ------------------- Markers for Coordinates -------------------
    
    for j in maps_t:
        
        timer = randint(8,15)
        #pause de loop
        print("coor "+str(timer))
        sleep(timer)
        
        if j == "NO ACCESS":
            coor.append("NO COORDINATES")
            continue
            
        resp = get(j)
        s = resp.text
        
        deb = s.find('setView')
        setview = s[deb:]
        fin = setview.find(']')
        setview_full =  setview[:fin+1]
        
        #Take out the elements 
        latLng = setview_full[setview_full.find('[')+1:setview_full.find(']')]
    
        coor.append(latLng)
        
    maps = maps + maps_t
        
    
    
    
    
    
    
    
    
    
    #############""
    z = z + 1
    
    
    
    
print(" !!! !!!! !!! DONE !!! !!! !!! ")  
    
```

```python
coor
```

# Join the DATA

```python
combo = []
for i,j,k,l in zip(pharmacies,liens,maps,coor):
    
    t = (i,j,k,l)
    combo.append(t)
```

## Convert into a pandas object

```python
import pandas as pd
```

```python
df = pd.DataFrame(combo,columns=['pharmacie','lien','access','coordonnee'])
```
