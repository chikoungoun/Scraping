# Telecontact Comparator

```python
import pickle
import pandas as pd
```

```python
df = pd.read_csv("pdg_casa_full.csv",delimiter=';',index_col=0,converters={'telephone': lambda x: str(x)})


#load the on duty list
duty = pickle.load(open("tc_pdg.pickle",'rb'))
```

```python
df.head(2)
```

```python
noms = []
adresses = []
quartiers = []
telephones = []
coors = []
liens = []

c = 0
for i, row_i in duty.iterrows():

    for j,row_j in df.iterrows():


        if(row_i['pharmacie'] == row_j['pharmacie']):
            c = c+1
            print("Yosha !!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!!! ")
            # Get the data from Duty
            #ph.append(df.loc[df['telephone'] == row_i['telephone']])
            noms.append(row_i['pharmacie'])
            quartiers.append(row_i['quartier'])
            telephones.append(row_j['telephone'])
            coors.append(row_j['coordonnee'])
            adresses.append(row_j['adresse'])
            liens.append(row_j['lien'])
        else:
            print('Lame')
            
            
print("\n *** *** Done *** ***")
print(c)
```

```python
combo = []
for i,j,k,l,m,n in zip(noms,quartiers,adresses,coors,telephones,liens):
    #Ajouter le lieu
    t = (i,j,k,l,m,n)
    combo.append(t)
```
