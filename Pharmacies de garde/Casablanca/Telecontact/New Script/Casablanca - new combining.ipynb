# Combining the files

```python
import pandas as pd
```

```python
ph1 = pd.read_csv('ph1.csv',delimiter=';',index_col=0)
```

```python
ph2 = pd.read_csv('ph2.csv',delimiter=';',index_col=0)
ph3 = pd.read_csv('ph3.csv',delimiter=';',index_col=0)
ph4 = pd.read_csv('ph4.csv',delimiter=';',index_col=0)
ph5 = pd.read_csv('ph5.csv',delimiter=';',index_col=0)
```

```python
ph6 = pd.read_csv('ph6.csv',delimiter=';',index_col=0)
ph7 = pd.read_csv('ph7.csv',delimiter=';',index_col=0)
ph8 = pd.read_csv('ph8.csv',delimiter=';',index_col=0)
ph9 = pd.read_csv('ph9.csv',delimiter=';',index_col=0)
```

```python
ph10 = pd.read_csv('ph10.csv',delimiter=';',index_col=0)
ph11 = pd.read_csv('ph11.csv',delimiter=';',index_col=0)
ph12 = pd.read_csv('ph12.csv',delimiter=';',index_col=0)
```

```python
ph13 = pd.read_csv('ph13.csv',delimiter=';',index_col=0)
ph14 = pd.read_csv('ph14.csv',delimiter=';',index_col=0)
ph15 = pd.read_csv('ph15.csv',delimiter=';',index_col=0)
```

```python
ph16 = pd.read_csv('ph16.csv',delimiter=';',index_col=0)
ph17 = pd.read_csv('ph17.csv',delimiter=';',index_col=0)
ph18 = pd.read_csv('ph18.csv',delimiter=';',index_col=0)
```

```python
frames = [ph1,ph2,ph3,ph4,ph5,ph6,ph7,ph8,ph9,ph10,ph11,ph12,ph13,ph14,ph15,ph16,ph17,ph18]
result = pd.concat(frames)
```

```python
result
```

### Taking out all the duplicates

```python
result.drop_duplicates(inplace=True)
```

### Sort by alphabetical order & resset index

```python
result.sort_values(by="pharmacie",inplace=True)
```

```python
result.reset_index(drop=True,inplace=True)
```

```python
result
```

### Export to CSV format

```python
result.to_csv("combi_ph1.csv",sep=";",header=True,encoding='utf-8')
```
