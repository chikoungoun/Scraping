# Scraping the telecontact page for all the Pharmacy in Casablanca

```python
import requests

url = 'https://www.telecontact.ma/liens/pharmacies/casablanca.php'
response = requests.get(url)
response
```

```python
response.text
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

```python
soup.prettify
```

# ---------------------------------------

### Clean the code and keep only what we need

```python
moodal = soup.find_all('a',class_='moodalbox')


pharma = []
links = []
for i in moodal:
    if 'pharmacie' in i['href'] :
        print(i.text)
        pharma.append(i.text)   
        links.append(i['href'])

```

** Get the links too **

```python
links
```

** Grouping the names in a list **

```python
pharmacies = []
for x,i in enumerate(pharma,1):
    print(x)
    if (x % 3 == 0) :
        print(i)
        pharmacies.append("Pharmacie"+i)
            
```

** Grouping the links in a list **

```python
liens = []
for x,i in enumerate(links,1):
    print(x)
    if (x % 3 == 0) :
        print(i)
        liens.append("https://www.telecontact.ma"+i)
```

### Scraping the LAT and LONG from each localization

```python
from time import time
from time import sleep
from random import randint

start_time = time()
requests = 0

from requests import get
```

```python
coor = []
for i in access:
    print(i)
    
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    resp = get(i)
    s = resp.text
    
    deb = s.find('setView')
    setview = s[deb:]
    fin = setview.find(']')
    setview_full =  setview[:fin+1]
    
    #Take out the elements 
    latLng = setview_full[setview_full.find('[')+1:setview_full.find(']')]
    
    coor.append(latLng)

print('*** *** *** DONE *** *** ***')
    
```

```python
print(coor)
print(len(coor))
```

## Join  the data

```python
combo = []
for i,j,k,l in zip(pharmacies,liens,access,coor):
    #Ajouter le lieu
    t = (i,j,k,l)
    combo.append(t)
```
