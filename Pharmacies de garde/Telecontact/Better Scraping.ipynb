```python
import requests

url = 'https://www.telecontact.ma/trouver/index.php?nxo=moteur&nxs=process&string=pharmacie&ou=casablanca&page=6&produit='
response = requests.get(url)
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

```python
results = soup.find_all('article',class_='drs col-xs-12')
```
