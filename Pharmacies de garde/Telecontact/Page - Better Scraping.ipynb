```python
import requests

url = 'https://www.telecontact.ma/trouver/index.php?nxo=moteur&nxs=process&string=pharmacie&ou=casablanca&page=6&produit='
response = requests.get(url)
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

** Isolate each block **

```python
results = soup.find_all('article',class_='drs col-xs-12')
```

## Scraping the Access Plans

```python
maps = []
for i in results:
    print(i.find('div',class_="col-xs-12 no-margin").find('div',class_="col-xs-12 no-margin").find('a'))
    m = i.find('div',class_="col-xs-12 no-margin").find('div',class_="col-xs-12 no-margin").find('a')
    
    if m['title'] == "Plan d'acc�s":
        maps.append("https://www.telecontact.ma"+m['href'])
    else:
        maps.append("NO ACCESS")
    print("  **************************************** ")
```

## Scraping the coordinates of all the possible links

```python
from time import time
from time import sleep
from random import randint

start_time = time()
requests = 0

from requests import get
```

```python
coor = []
for i in maps:
    print(i)
    
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    if i == "NO ACCESS":
        coor.append("NO COORDINATES")
        continue
    
    resp = get(i)
    s = resp.text
    
    deb = s.find('setView')
    setview = s[deb:]
    fin = setview.find(']')
    setview_full =  setview[:fin+1]
    
    #Take out the elements 
    latLng = setview_full[setview_full.find('[')+1:setview_full.find(']')]
    
    coor.append(latLng)

print('*** *** *** DONE *** *** ***')
```
