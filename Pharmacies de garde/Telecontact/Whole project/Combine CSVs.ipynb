# Combining all the CSVs 

Because the page results are mostly random, I had to run the code several times, hoping to gather most of the data in seperate chunks and then link them all together after taking out all the duplicates

```python
import pandas as pd
```

```python
ph1 = pd.read_csv('ph1.csv',delimiter=';',index_col=0)
```

```python
ph2 = pd.read_csv('ph2.csv',delimiter=';',index_col=0)
ph3 = pd.read_csv('ph3.csv',delimiter=';',index_col=0)
ph4 = pd.read_csv('ph4.csv',delimiter=';',index_col=0)
ph5 = pd.read_csv('ph5.csv',delimiter=';',index_col=0)
```

```python
ph6 = pd.read_csv('ph6_40.csv',delimiter=';',index_col=0)
ph7 = pd.read_csv('ph7_10.csv',delimiter=';',index_col=0)
ph8 = pd.read_csv('ph8_20.csv',delimiter=';',index_col=0)
ph9 = pd.read_csv('ph9_15.csv',delimiter=';',index_col=0)
```

```python
ph10 = pd.read_csv('ph10_15.csv',delimiter=';',index_col=0)
ph11 = pd.read_csv('ph11_10.csv',delimiter=';',index_col=0)
ph12 = pd.read_csv('ph12_10.csv',delimiter=';',index_col=0)
ph13 = pd.read_csv('ph13_20.csv',delimiter=';',index_col=0)
```

```python
ph14 = pd.read_csv('ph14_20.csv',delimiter=';',index_col=0)
ph15 = pd.read_csv('ph15_40.csv',delimiter=';',index_col=0)
```

### Concatenate the data

```python
frames = [ph1,ph2,ph3,ph4,ph5,ph6,ph7,ph8,ph9,ph10,ph11,ph12,ph13,ph14,ph15]
result = pd.concat(frames)
```
