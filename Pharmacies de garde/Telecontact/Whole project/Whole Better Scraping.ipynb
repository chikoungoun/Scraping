```python
url ="https://www.telecontact.ma/liens/pharmacies/casablanca.php&page="
```

```python
from bs4 import BeautifulSoup
```

```python
from time import time
from time import sleep
from random import randint
```

```python
start_time = time()
requests = 0
```

```python
from requests import get
```

```python
liens = []
pharmacies = []
maps = []

z = 1
while (z< 10):
    print(" --- ---  PAGE"+z+" --- ---")
    print(url+str(z))
    lien = url+str(z)
    
    response = get(lien)
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    soup = BeautifulSoup(response.text,'html.parser')
    
    results = soup.find_all('article',class_='drs col-xs-12')
    
    for i in results:
        liens.append("https://www.telecontact.ma"+str(i.find('span',id='resultats_h3_span').findAll('a')[1]['href']))
        
        
    for i in results:
        pharmacies.append("Pharmacie "+str(i.find('span',id='resultats_h3_span').findAll('a')[2].text))
        
        
    for i in results:
        m = i.find('div',class_="col-xs-12 no-margin").find('div',class_="col-xs-12 no-margin").find('a')
    
        if m['title'] == "Plan d'acc�s":
            maps.append("https://www.telecontact.ma"+m['href'])
        else:
            maps.append("NO ACCESS")
    
    # ------------------- Markers for Coordinates -------------------
    
    
    
    
    
    
    
    
    
    #############""
    z = z + 1
    
    
    
    
print(" !!! !!!! !!! DONE !!! !!! !!! ")  
    
```

# Join the DATA

```python
combo = []
for i,j,k in zip(pharmacies,liens,maps):
    
    t = (i,j,k)
    combo.append(t)
```
