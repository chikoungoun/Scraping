```python
import requests

url = 'https://www.med.tn/pharmacie/tunis'
response = requests.get(url)
response
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

```python
soup
```

### Get all the pharmacie cards

```python
results = soup.find_all('div',class_="searchResults-itemDoctor")
```

```python
print(len(results))
results
```

### Scrape the names

```python
noms = []
for i in results:
    print(i.find('span',class_='practitioner-name').text)
    noms.append((i.find('span',class_='practitioner-name').text).lower())
```

```python
noms
```

### Scrape the adresses

```python
pre_adresses = []
for i in results:
    print((i.find('div',class_='practitioner-address').find('p').text).strip())
    pre_adresses.append((i.find('div',class_='practitioner-address').find('p').text).strip())
```

```python
pre_adresses
```

cleaning up the space at the end

```python
adresses = []
for i in pre_adresses:
    adresses.append(i.replace('Tunisie','').replace('  ',''))
```

```python
adresses
```

### Scrape the phones

```python
phones = []
for i in results:
    print(i.find('span',class_='view-tel').text)
    phones.append(i.find('span',class_='view-tel').text)
```

### Scrape the coordinates

```python
pre_coors = []
for i in results:
    print(i.find('a',class_='btn').findNext('a')['href'])
    pre_coors.append(i.find('a',class_='btn').findNext('a')['href'])
```

taking out the wanted coordinates

```python
coors = []
for i in pre_coors:
    eq = i.find('=')
    print(i[eq+1:])
    coors.append(i[eq+1:])
```

```python
coors
```

## Merge the data we got

```python
garde = []
for i,j,k,l in zip(noms,adresses,coors,phones):
    t = (i,j,k,l)
    garde.append(t)
```

```python
garde
```

## Convert into Pandas object

```python
import pandas as pd
```

```python
df = pd.DataFrame(garde,columns=['nom','adresse','coordonnee','telephone'])
```
