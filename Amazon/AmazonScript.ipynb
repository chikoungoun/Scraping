# Scraping sur l'ensemble des données de la recherche Amazon : 'Web Scraping"

```python
from requests import get
from bs4 import BeautifulSoup
import numpy as np
import pandas as pd
from random import randint
from time import sleep
from time import time
from IPython.core.display import clear_output
from warnings import warn
```

```python
response = get('https://www.amazon.fr/s/ref=sr_pg_2?rh=i%3Aaps%2Ck%3Aweb+scraping&page='+str(1)+'&keywords=web+scraping&ie=UTF8&qid=1527705889')
```

```python
response
```

```python
noms = []
dates = []
auteurs = []
prix = []

pages = np.arange(1,4)

# Preparing the monitoring of the loop
start_time = time()
requests = 0


for page in pages:
    
    response = get('https://www.amazon.fr/s/ref=sr_pg_2?rh=i%3Aaps%2Ck%3Aweb+scraping&page='+ str(page) +'&keywords=web+scraping&ie=UTF8&qid=1527705889')
    
    #Faire une pause pour pas trop déconner
    sleep(randint(6,10))
    
    # Monitor the requests
    requests += 1
    elapsed_time = time() - start_time
    print('Request:{}; Frequency: {} requests/s'.format(requests, requests/elapsed_time))
    clear_output(wait = True)
    
     # Throw a warning for non-200 status codes
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
        
    soup = BeautifulSoup(response.text,'html.parser')   
    
    amazon_container = soup.find('div',id='atfResults').find_all('li')

    type(amazon_container)

    for con in amazon_container:
    
        #Nom   
        nom = con.h2.text
        noms.append(nom)
    
        #Date
        date = con.find('span',class_='a-size-small a-color-secondary').text
        dates.append(date)
    
        #Auteurs
        x = con.find_all('div',class_='a-row a-spacing-none')
        try:
            span = x[1].span
        except IndexError:
            auteurs.append(np.nan)
        if span is not None:
            #print(span.nextSibling.text)
            auteurs.append(span.nextSibling.text)
        else:
            auteurs.append(np.nan)
        
    
        #Prix 
        if con.find('span',class_='a-size-base a-color-price s-price a-text-bold') is not None:
            prix.append(con.find('span',class_='a-size-base a-color-price s-price a-text-bold').text)
        else:
            prix.append(np.nan)
    
    
    
    



        

    

        
        
print('Done')
```
