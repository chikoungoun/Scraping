**Import Selenium to open the links through the Chrome browser**

```python
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')# access the browser in incognito mode
options.add_argument('--headless') # access the browser without having to open it
```

```python
driver = webdriver.Chrome("C:/webdrivers/chromedriver85.exe", options=options)
```

```python
from selenium.common.exceptions import NoSuchElementException 
```

```python
if driver.find_elements_by_css_selector('.orangeTit'):
    print ("Element exists")
```

```python
houses = []

for i in links:
    driver.get(i)
    
    print(i)
    
    #Valeur par dfaut
    superficie ="NO superf."
    pieces = "NO pieces"
    etage = " NO etage"
    
    
    # Test si l'lment existe
    if driver.find_elements_by_css_selector('.orangeTit'):
        # prix
        prix = driver.find_element_by_class_name('orangeTit')
    
        #titre
        titre = driver.find_element_by_class_name('greyTit')
    
        #tags
        tags = driver.find_elements_by_class_name('tagProp')
    
        for j in tags:
            if 'm' in j.text:
                print('Hourra metre')
                print(j.text)
                superficie = j.text
                
            if 'Pices' in j.text:
                print('Hourra piece')
                print(j.text)
                pieces = j.text
                
            if 'tage' in j.text:
                print('Hourra etage')
                print(j.text)
                etage = j.text
        
        t = (titre.text,prix.text,superficie,pieces,etage,i)
        print(t)
        houses.append(t)
        
    else:
        print('element does not exist')
    
    
    try:
        driver.find_element_by_class_name('orangeTit')
    except NoSuchElementException:
        print('prout')

   
    
    
   
    
    
print(' ###\n  DONE \n ###  !  ')
    
    
    
```
