```python
import requests

from bs4 import BeautifulSoup
```

```python
from time import time
from time import sleep
from random import randint

start_time = time()

```

While analyzing the list rows, we realize that they are split in 3 categories : **Spremium, Premium and Basic**

```python
spremium = []
premium = []
basic = []
```

## Scraping all the links

```python
c = 1
while c < 18:
    
    url = 'https://www.mubawab.ma/fr/ct/casablanca/immobilier-a-vendre-all:prmn:%3E1000000:prmx:%3C2000000:esmn:%3E80:esmx:%3C130:et:y:sc:apartments-for-sale:p:'+str(c)
    response = requests.get(url)
    print(response)
    print("### \n ###   Page "+str(c)+"  ### \n ### \n")
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    soup = BeautifulSoup(response.text,'html.parser')
    
    
    #Bloc list
    ulist = soup.find("ul",class_="ulListing")
    
    #Spremium
    islist = ulist.find_all("li",class_="SpremiumBox w100")
    for i in islist:
        print(i['linkref'])
        spremium.append(i['linkref'])
        
        print(" ***\n ***\n ***\n ***\n")
        
    #Premium
    ilist = ulist.find_all("li",class_="premiumBox w100")
    for i in ilist:
        print(i['linkref'])
        premium.append(i['linkref'])
        
    #BasicList
    basicList = ulist.find_all("li",class_="basicList")
    for i in basicList:
        print(i['linkref'])
        basic.append(i['linkref'])
        
    
    
    
    #Increment
    c = c + 1
    
    
    
print(' ### !!! ###   DONE   ### !!! ### ')
```
