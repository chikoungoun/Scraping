```python
from selenium import webdriver
```

```python
options = webdriver.ChromeOptions()
options.add_argument('--ignore-certificate-errors')
options.add_argument('--incognito')# access the browser in incognito mode
options.add_argument('--headless') # access the browser without having to open it
```

```python
driver = webdriver.Chrome("C:/webdrivers/chromedriver.exe", options=options)
```

```python
base = "https://www.fifatrainingcentre.com"
driver.get('https://www.fifatrainingcentre.com/en/resources-tools/football-language')
page = driver.page_source
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(page,'html.parser')
type(soup)
```

```python
soup
```

```python
root_level_nav = soup.find('ul',class_='first-level')
root_level_list = root_level_nav.find_all('a')
```

```python
root_level = []

for i in root_level_list:
    root_level.append((base+i['href'],1))
    
root_level
    
```

**I am going to classify the data I have in terms of levels first. Then reorder the whole thing in a tree data structure.**
