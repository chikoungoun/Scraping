# Scraping Cine Atlas : génération JSON

```python
import re
import json
import requests
from ast import literal_eval

url = 'https://www.cineatlas.com/'

html_data = requests.get(url).text
movieData = re.findall(r'movieData = ({.*?}), movieDataByReleaseDate', html_data, flags=re.DOTALL)[0]
movieData = re.sub(r'\s*/\*.*?\*/\s*', '', movieData)   # remove comments
movieData = literal_eval(movieData) # in movieData you have now the information about the current movies

print(json.dumps(movieData, indent=4))  # print data to the screen
```

```python
jfile = json.dumps(movieData, indent=4)
```

```python
data = json.loads(jfile)
```

```python
dates = [ ]
for k in data:
    dates.append(k)
    print(k)
```

```python
liens = []
noms = []
images = []
heures = []
heure_str = []
```

```python
for i in data.get(dates[0]):
    print(i['url'])
    liens.append("https://www.cineatlas.com/movie/"+str(i['url']))
```

```python
for i in data.get(dates[0]):
    print(i['title'])
    noms.append((i['title']))
```

```python
for i in data.get(dates[0]):
    print(i['image-portrait'])
    images.append((i['image-portrait']))
```

```python
for i in data.get(dates[0]):
    h = []
    for j in i['times']:
        print(j['time'])
        h.append(j['time'])
    heures.append(h)
```

```python
heures
```

```python
for i in heures:
    s = ""
    for j in i:
        s =s+str(j)+" "
    heure_str.append(s)
```

```python
from time import time
from time import sleep
from random import randint
```

```python
start_time = time()
requests = 0
```

```python
from bs4 import BeautifulSoup
from requests import get
```

```python
descs = []
for i in liens:
    print(i)
    response = get(i)
    
    timer = randint(8,15)
    #pause de loop
    print(timer)
    sleep(timer)
    
    #renvoyer un warning pour les non 'status code : 200'
    if response.status_code != 200:
        warn('Request: {}; Status code: {}'.format(requests, response.status_code))
            
    #parser en BS
    page_html = BeautifulSoup(response.text,'html.parser')  
    
    events = page_html.find('div',class_='mobile-hide movie-details-container')
    
    if events is None:
        descs.append("NO DESCRIPTION")
        continue
    
    event = events.find('div',class_='movie-details')   
    
        
    #dans le trou du cul du code (aka 4eme <p> du <div> cible )
    print(event.findAll('p',recursive=False)[3].text)
    
    descs.append(event.findAll('p',recursive=False)[3].text)
    
print("** Done **")  
```

```python
descs
```

## Tout regrouper en une seule structure

```python
prog = []
for i,j,k,l,m in zip(noms,heure_str,images,liens,descs):
    #Ajouter le lieu
    t = (i,dates[0],j,k,l,m,"Cine Atlas")
    prog.append(t)
```
