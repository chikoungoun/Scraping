# Projet de Scraping des évènements culturels au maroc : Ticket.ma 

## Partie  Thêatre - Humour

```python
import requests
```

```python
url_concerts = 'https://www.ticket.ma/ticket/theatre-humour'
response = requests.get(url_concerts)
response
```

```python
print(response.text[:500])
```

```python
from bs4 import BeautifulSoup

soup = BeautifulSoup(response.text,'html.parser')
type(soup)
```

```python
spectacles = soup.find('div',{"class":"home clearfix local-events"})
```

```python
spectacles = spectacles.find('ul')
```

```python
spectacles
```

## Scraping des liens

```python
liens = []
```

```python
for a in spectacles.find_all('a',{"class":"data-link"}):
    liens.append("https://www.ticket.ma"+str(a['href']))
```

```python
liens
```

### Scraping des dates (format plus intéressant sur la page d'accueil)

```python
spectacles.find_all('p',{"class":"event-date"})
```

```python
d = []
for i in spectacles.find_all('p',{"class":"event-date"}):
    d.append(i.get_text().replace("\t","").replace("\n",""))
```

```python
dates = []
for i in d:
    a,b = i.split('·')
    dates.append(a)
```

```python
dates
```

## Appliquer la Fonction Ticket

```python
def ticket(lien,date):
    url = lien
    response = requests.get(url)
    print(response)
    
    soup = BeautifulSoup(response.text,'html.parser')
    type(soup)
    
    #focus sur le cadre de travail
    event = soup.find('article',{'class':'event'})
    
    #extraire le nom
    nom = event.find('div',{'class':'event-titles'}).get_text().replace('\n','').replace('\t','')
    
    #extraire la localisation
    lieu = event.find('a',{'class':'fancybox map-location'}).get_text()
    
    #extraire l'heure du début du spectacle
    heure = event.find('span',{'class':'start dtstart'}).get_text()
    heure = heure[-5:]
    
    #extraire le lien de l'image
    img = event.find('div',{'id':'artist-image'}).find('img')
    img['src']
    image = "https://www.ticket.ma/"+str(img['src'])
    
    #retourner sous forme de tuple (nom, localisation, heure)
    return (nom,lieu,date,heure,image)
```

```python
len(liens)
```

```python
tickets = []    
i=0
while i < (len(liens)-1):
    tickets.append(ticket(liens[i],dates[i]))
    print(tickets[i])
    i+=1
```
